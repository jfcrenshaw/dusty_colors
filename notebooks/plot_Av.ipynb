```python
fig, ax = plt.subplots(figsize=(3, 3), dpi=200)

results = np.load(f"../results/stacks/{stack}/stack_fcolors.npz")
dp1_x = results["g-z_bin_centers"]
dp1_y = color_excess_to_av(results["g-z_avg"], "g-z", foreground_redshift=0.36)
dp1_err = color_excess_to_av(results["g-z_err"], "g-z", foreground_redshift=0.36)

ax.errorbar(
    dp1_x,
    dp1_y,
    dp1_err,
    ls="",
    marker="s",
    markersize=3,
    label="DP1",
    c="r",
    zorder=100,
)

ax.errorbar(
    menard_x,
    menard_y,
    menard_err,
    ls="",
    marker=".",
    label="Menard 2010",
    mfc="w",
    c="darkgreen",
)

ax.errorbar(
    kids_x,
    kids_y,
    kids_err,
    ls="",
    marker=".",
    label="Genc 2025",
    mfc="w",
    c="gray",
)
ax.errorbar(
    des_x,
    des_y,
    des_err,
    ls="",
    marker=".",
    label="McCleary 2026",
    mfc="w",
    c="darkblue",
)

ax.set(
    xscale="log",
    yscale="log",
    xlabel=r"$r_\perp$ [kpc]",
    ylabel=r"$A_V$ [mag]",
    xlim=(10, 1e4),
    ylim=(1e-5, 1),
)
ax.set_box_aspect(1)
ax.legend(loc="upper right", fontsize=7, frameon=False)

add_angular_axis(ax)


# Fit a power law at <100 kpc
def power_law(x, A, alpha):
    return A * (x / 20) ** alpha


mask = dp1_x < 100
popt, pcov = curve_fit(
    power_law,
    dp1_x[mask],
    dp1_y[mask],
    p0=[0.003, -0.86],
    sigma=dp1_err[mask],
    absolute_sigma=True,
)
print("Best fit power law:")
print(
    f"    scale = {popt[0]:.1e} ± {np.sqrt(pcov[0, 0]):.1e}; snr {popt[0] / np.sqrt(pcov[0, 0]):.2f}"
)
print(f"    index = {popt[1]:.2f} ± {np.sqrt(pcov[1, 1]):.2f}")

if save_figs:
    fig.savefig("../figures/fig_result_compare.pdf", bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(3, 3), dpi=200)

results = np.load(f"../results/stacks/{stack}/stack_fcolors.npz")
x = results["g-z_bin_centers"]
samples = results["g-z_jackknife_samples"]

for i, y in enumerate(samples):
    ls = {0: "-", 1: "--", 2: ":"}[i % 3]
    ax.plot(x, y, marker=".", ls=ls, label=f"{i+1}", c=f"C{i // 3}")

ax.set(
    xscale="log",
    yscale="log",
    xlabel=r"$r_\perp$ [kpc]",
    ylabel=r"$A_V$ [mag]",
    xlim=(10, 1e3),
    ylim=(1e-3, 1),
)
ax.set_box_aspect(1)

N_COLS = 3


def transpose_legend(items, ncols):
    return list(itertools.chain(*[items[i::ncols] for i in range(ncols)]))


handles, labels = ax.get_legend_handles_labels()
ordered_handles = transpose_legend(handles, N_COLS)
ordered_labels = transpose_legend(labels, N_COLS)
ax.legend(
    ordered_handles,
    ordered_labels,
    ncols=N_COLS,
    title="Jackknife region",
    handlelength=1.5,
    handletextpad=0.1,
    columnspacing=0.75,
    frameon=False,
)

add_angular_axis(ax)

if save_figs:
    fig.savefig("../figures/fig_result_jackknife.pdf", bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(3, 3), dpi=200)

results = np.load(f"../results/stacks/{stack}/stack_fcolors.npz")

for i, color in enumerate(["g-r", "r-i", "i-z"]):
    x = results[f"{color}_bin_centers"]
    y = color_excess_to_av(results[f"{color}_avg"], color, foreground_redshift=0.36)
    err = color_excess_to_av(results[f"{color}_err"], color, foreground_redshift=0.36)

    if i == 0:
        offset = 1
    elif i == 1:
        offset = 0.9
    elif i == 2:
        offset = 1.1
    ax.errorbar(
        x * offset,
        y,
        err,
        ls="",
        marker="s",
        markersize=3,
        label=f"${color}$",
        zorder=100,
    )

ax.set(
    xscale="log",
    yscale="log",
    xlabel=r"$r_\perp$ [kpc]",
    ylabel=r"$A_V$ [mag]",
    xlim=(10, 1e3),
    ylim=(1e-3, 1),
)
ax.set_box_aspect(1)

ax.legend(loc="upper right", fontsize=7, frameon=False)

add_angular_axis(ax)

if save_figs:
    fig.savefig("../figures/fig_result_colors.pdf", bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(4, 4), dpi=200)


def plot(stack, label, offset=1.0, marker="s"):
    results = np.load(f"../results/stacks/{stack}/stack_fcolors.npz")
    dp1_x = results["g-z_bin_centers"]
    dp1_y = color_excess_to_av(results["g-z_avg"], "g-z", foreground_redshift=0.36)
    dp1_err = color_excess_to_av(results["g-z_err"], "g-z", foreground_redshift=0.36)

    ax.errorbar(
        dp1_x * offset,
        dp1_y,
        dp1_err,
        ls="",
        marker=marker,
        markersize=4 if label == "Default" else 3,
        label=label,
        mfc="w" if label == "Default" else None,
        zorder=-1 if label == "Default" else 10,
    )


plot("dp1_default", "Default")
plot("dp1_uniform_random", "Uniform randoms", 0.97, marker="x")
plot("dp1_no_random", "No random correction", 0.95, marker="x")
plot("dp1_no_flip", "No flipped correction", 1.05, marker="x")
plot("dp1_smaller_ap", "Smaller aperture", 0.90, marker=".")
plot("dp1_larger_ap", "Larger aperture", 1.10, marker=".")
plot("dp1_smaller_dz", "$\Delta z = 0.1$", 0.85, marker="^")
plot("dp1_larger_dz", "$\Delta z = 0.3$", 1.15, marker="^")


ax.set(
    xscale="log",
    yscale="log",
    xlabel=r"$r_\perp$ [kpc]",
    ylabel=r"$A_V$ [mag]",
    xlim=(10, 1e3),
    ylim=(1e-3, 1),
)
ax.set_box_aspect(1)
ax.legend(loc="upper right", fontsize=7, frameon=False)

add_angular_axis(ax)

if save_figs:
    fig.savefig("../figures/fig_result_sensitivity.pdf", bbox_inches="tight")
```
